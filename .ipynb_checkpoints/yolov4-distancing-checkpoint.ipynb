```python

def is_close(p1, p2):
    """
    #================================================================
    # 1. Purpose : Calculate Euclidean Distance between two points
    #================================================================    
    :param:
    p1, p2 = two points for calculating Euclidean Distance
    :return:
    dst = Euclidean Distance between two 2d points
    """
    dst = math.sqrt(p1**2 + p2**2)
    #=================================================================#
    return dst 
```

```python
import cv2
import numpy as np
net = cv2.dnn.readNetFromDarknet("yolov3.cfg","yolov3_1100.weights")
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
classes = [line.strip() for line in open("obj.names")]
colors = [(0,0,255),(255,0,0),(0,255,0)]#

def yolo_detect(frame):
    # forward propogation
    img = cv2.resize(frame, None, fx=0.4, fy=0.4)
    height, width, channels = img.shape 
    blob = cv2.dnn.blobFromImage(img, 1/255.0, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    # get detection boxes
    class_ids = []
    confidences = []
    boxes = []
    centroid_dict = dict()#test
    objectId = 0#test
    for out in outs:
        for detection in out:
            tx, ty, tw, th, confidence = detection[0:5]
            scores = detection[5:]
            class_id = np.argmax(scores)  
            if confidence > 0.5:#原先設定0.3   
                center_x = int(tx * width)
                center_y = int(ty * height)
                w = int(tw * width)
                h = int(th * height)

                # 取得箱子方框座標
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                x_max = int(center_x + w / 2)
                y_max = int(center_y + h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
                centroid_dict[objectId] = (int(x), int(y), x, y, x_max, y_max)#test
                objectId += 1#test
                
    # draw boxes
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.3, 0.4)
    font = cv2.FONT_HERSHEY_PLAIN
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            color = colors[class_ids[i]]
            cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
 #           cv2.putText(img, label, (x, y -5), font, 3, color, 3)
            
    red_zone_list = [] # List containing which Object id is in under threshold distance condition. 
    red_line_list = []
    for (id1, p1), (id2, p2) in combinations(centroid_dict.items(), 2): # 獲取所有的detections組合, #List of multiple items - id1 1, points 2, 1,3
            dx, dy = p1[0] - p2[0], p1[1] - p2[1]  # Check the difference between centroid x: 0, y :1
            distance = is_close(dx, dy) # 歐式距離
            if distance < 100:# 設定社交距離
                if id1 not in red_zone_list:
                    red_zone_list.append(id1)       
                    red_line_list.append(p1[0:2])   
                if id2 not in red_zone_list:
                    red_zone_list.append(id2)		
                    red_line_list.append(p2[0:2])
    for idx, box in centroid_dict.items():  # dict (1(key):red(value), 2 blue)  idx - key  box - value
        if idx in red_zone_list and idx in indexes:   # if id is in red zone list
            #cv2.rectangle(img, (box[2], box[3]), (box[4], box[5]), (0, 0, 255), 2)# Create Red bounding boxes  #starting point, ending point size of 2
            cv2.putText(img, "too close", (50,20), font, 3, (0, 0, 255), 2)
    
    for check in range(0, len(red_line_list)-1):# 畫線
        start_point = red_line_list[check] 
        end_point = red_line_list[check+1]
        check_line_x = abs(end_point[0] - start_point[0])# 計算x座標 
        check_line_y = abs(end_point[1] - start_point[1])# 計算y座標
        if (check_line_x < 75) and (check_line_y < 25):
            cv2.line(img, start_point, end_point, (0, 0, 255), 2) 
           
		#=================================================================#
      
    
    return img
```
